```python
import re
from nltk.stem import WordNetLemmatizer
import pandas as pd
import nltk
nltk.download('wordnet')


lemmatizer = WordNetLemmatizer()
corpus = []

# Read the data from spam.csv
try:
    df = pd.read_csv('spam.csv', encoding='latin-1')
    messages = df['v2'].tolist() # Assuming 'v2' is the column with the messages
except FileNotFoundError:
    print("Error: spam.csv not found. Please make sure the file was uploaded and extracted correctly.")
    messages = [] # Initialize messages as an empty list to avoid further errors


for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages[i])
  review=review.lower()
  review=review.split()

  review=[lemmatizer.lemmatize(word)for word in review]
  # Join the list of words back into a string.
  review=' '.join(review)
  corpus.append(review)
```
